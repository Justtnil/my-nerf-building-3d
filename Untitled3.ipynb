```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# 🧠 Simple NeRF Model - Your AI Brain!
class SimpleNeRF(nn.Module):
    def __init__(self):
        super(SimpleNeRF, self).__init__()

        print("🧠 Building your NeRF brain...")

        # The neural network that learns 3D scenes
        self.network = nn.Sequential(
            nn.Linear(3, 128),    # Input: 3D coordinates (x, y, z)
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, 4)     # Output: RGB color + density
        )

    def forward(self, positions):
        # Magic happens here: 3D position → Color & Density
        return self.network(positions)

# 🎯 Create your NeRF model
print("🚀 Creating your NeRF model...")
model = SimpleNeRF()

# 🔢 Model statistics
total_params = sum(p.numel() for p in model.parameters())
print(f"📊 Your NeRF has {total_params:,} parameters!")
print(f"🎯 Input: 3D coordinates (x, y, z)")
print(f"🎨 Output: Color (R, G, B) + Density")

# 🧪 Test your NeRF with dummy 3D points
test_points = torch.randn(100, 3)  # 100 random 3D points
with torch.no_grad():
    output = model(test_points)
    colors = output[:, :3]  # RGB colors
    density = output[:, 3]  # Density values

print(f"✅ NeRF test successful!")
print(f"📈 Output shape: {output.shape}")
print(f"🌈 Color range: {colors.min():.2f} to {colors.max():.2f}")
print(f"💨 Density range: {density.min():.2f} to {density.max():.2f}")

print("\n🎉 Your NeRF model is ready!")
```

```python
import torch.optim as optim
from PIL import Image
import torchvision.transforms as transforms

print("🎓 Starting NeRF training on your building photos!")

# 📸 Load and prepare your building photos
data_dir = pathlib.Path("data/building1/images")
photos = list(data_dir.glob("*.jpg"))

# Convert photos to tensors
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Small size for fast training
    transforms.ToTensor()
])

images = []
for photo_path in photos:
    img = Image.open(photo_path)
    img_tensor = transform(img)
    images.append(img_tensor)

print(f"📊 Loaded {len(images)} building photos")
print(f"📏 Image size: {images[0].shape}")

# 🎯 Set up training
optimizer = optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

# 🏃‍♀️ Mini training session (simplified for demo)
print("\n🚀 Training your NeRF...")
print("⚡ Running 100 training steps...")

training_losses = []

for step in range(100):
    # Generate random 3D points in the scene
    random_points = torch.randn(256, 3)

    # Predict colors and density
    predictions = model(random_points)

    # Simple training target (in real NeRF, this comes from ray tracing)
    target_colors = torch.rand(256, 3)  # Random colors for demo
    target_density = torch.rand(256, 1)  # Random density for demo
    target = torch.cat([target_colors, target_density], dim=1)

    # Calculate loss and update
    loss = loss_fn(predictions, target)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    training_losses.append(loss.item())

    if step % 20 == 0:
        print(f"Step {step:3d}: Loss = {loss.item():.4f}")

print(f"\n✅ Training complete!")
print(f"📉 Final loss: {training_losses[-1]:.4f}")

# 📊 Show training progress
plt.figure(figsize=(8, 4))
plt.plot(training_losses)
plt.title("NeRF Training Progress")
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

print("🎉 Your NeRF has learned from the building photos!")
```

```python
# 🧪 Test your trained NeRF model!
print("🧪 Testing your trained NeRF model...")

# Generate a 3D grid of points (like a 3D scanner)
x = torch.linspace(-1, 1, 20)
y = torch.linspace(-1, 1, 20)
z = torch.linspace(-1, 1, 20)

# Create 3D grid
xx, yy, zz = torch.meshgrid(x, y, z, indexing='ij')
test_points = torch.stack([xx.flatten(), yy.flatten(), zz.flatten()], dim=1)

print(f"🔍 Testing {test_points.shape[0]} 3D points...")

# Get NeRF predictions
with torch.no_grad():
    predictions = model(test_points)
    colors = predictions[:, :3]  # RGB colors
    densities = predictions[:, 3]  # Density values

print(f"✅ NeRF predictions complete!")
print(f"🌈 Color range: {colors.min():.3f} to {colors.max():.3f}")
print(f"💨 Density range: {densities.min():.3f} to {densities.max():.3f}")

# 📊 Visualize the density field (3D structure)
densities_3d = densities.reshape(20, 20, 20)

# Show a 2D slice of the 3D density field
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(densities_3d[:, :, 10], cmap='viridis')
plt.title("NeRF Density Field (XY Slice)")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label='Density')

plt.subplot(1, 2, 2)
plt.imshow(densities_3d[:, 10, :], cmap='viridis')
plt.title("NeRF Density Field (XZ Slice)")
plt.xlabel("X")
plt.ylabel("Z")
plt.colorbar(label='Density')

plt.tight_layout()
plt.show()

print("🎉 Your NeRF has created a 3D understanding of the building!")
print("💡 The bright areas show where your NeRF thinks solid building parts are!")
```
